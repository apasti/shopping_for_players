```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
import warnings

#settings
warnings.filterwarnings("ignore")
```

```python
def data_selection(position):
    df = pd.read_csv("../raw_data/data_cleaned.csv")
    df_selected = df[df['position']==position]
    return df_selected
```

```python
def data_scaling(df):
    minmaxscaler = MinMaxScaler()
    df['highest_market_value_in_eur']= minmaxscaler.fit_transform(df[['highest_market_value_in_eur']])
    return df
```

```python
def data_encoding(df):
    ohe = OneHotEncoder(sparse=False)
    encoder =  OrdinalEncoder()
    df[ohe.get_feature_names_out()] = ohe.fit_transform(df[['foot']])
    df['country_of_citizenship'] = encoder.fit_transform(df[['country_of_citizenship']])
    return df
```

```python
def drop_columns(df):
    df.drop(columns=['Unnamed: 0','foot','position'], inplace = True)
    return df
```

```python
def data_preprocess(df):
    df = data_scaling(df)
    df = data_encoding(df)
    df = drop_columns(df)
    return df
```

```python
df = data_selection('Attack')
data_preprocess(df)
```
